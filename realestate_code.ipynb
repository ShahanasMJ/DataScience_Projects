```python

# Calculate Q1 and Q3
Q1 = df['lotarea'].quantile(0.25)
Q3 = df['lotarea'].quantile(0.75)

# Calculate IQR
IQR = Q3 - Q1

# Define bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f"Lower Bound: {lower_bound}, Upper Bound: {upper_bound}")

# Remove outliers
df_cleaned = df[(df['lotarea'] >= lower_bound) & (df['lotarea'] <= upper_bound)]

# Display cleaned DataFrame
print(df_cleaned)

# Boxplot of cleaned data
plt.boxplot(df_cleaned['Values'])
plt.title('Boxplot of Cleaned Values')
plt.ylabel('Values')
plt.show()
```

```python
train_set = train_sett.copy()
test_set = test_sett.copy()
 
y_train = train_set['saleprice']
y_test  = test_set['saleprice']
 
X_train = train_set.drop('saleprice', axis=1).reset_index(drop=True)
X_test  = test_set.drop('saleprice', axis=1).reset_index(drop=True)
 
y_train.index = X_train.index
y_test.index  = X_test.index
```

```python
from sklearn.base              import BaseEstimator, TransformerMixin
from sklearn.preprocessing     import OneHotEncoder, StandardScaler
```

```python
numerical   = list()
categorical = list()
 
[numerical.append(feature) if df[feature].dtype != 'object' else categorical.append(feature) for feature in df.columns]
numerical.remove('saleprice')
 
print(categorical)
print(numerical)
```

```python
loe = LeaveOneOutEncoder(cols=categorical)
X_train = loe.fit_transform(X_train, y_train)
X_test  = loe.transform(X_test)
```

```python
X_train
```

```python
#scaling data

columns = X_train.columns
# create a StandardScaler object
std = StandardScaler()

# fit and transform the data
df_encoded = std.fit_transform(X_train[columns])
# create a new DataFrame with the scaled data
X_train    = pd.DataFrame(data=df_encoded, columns=std.get_feature_names_out())
 
df_encoded = std.transform(X_test[columns])
X_test     = pd.DataFrame(data=df_encoded, columns=std.get_feature_names_out())
 
X_train.head(3)
```

```python
#machine learning
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.linear_model import Ridge, Lasso, ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import AdaBoostRegressor, ExtraTreesRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
models = [
    LinearRegression(),
    Ridge(),
    Lasso(),
    ElasticNet(),
    DecisionTreeRegressor(),
    RandomForestRegressor(),
    GradientBoostingRegressor(),
    SVR(),
    KNeighborsRegressor(),
    AdaBoostRegressor(),
    ExtraTreesRegressor(),
    XGBRegressor()
]
 
for model in models:
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
 
    print(f"{model.__class__.__name__} - MSE: {mse}, R^2: {r2}")
```

```python
# Parameter grids for hyperparameter tuning [fine tune method]
param_grid ={
   "GradientBoostingRegressor": {
    "n_estimators": [50, 100, 200],
    "learning_rate": [0.01, 0.1, 0.2],
    "max_depth": [3, 4, 5]
  }
  
}
```

```python
#grid search

from sklearn.model_selection import GridSearchCV
 
# Example for hyperparameter tuning with GridSearchCV
param_grid = param_grid['GradientBoostingRegressor']
 
grid_search = GridSearchCV(GradientBoostingRegressor(), param_grid, cv=5, scoring='neg_mean_squared_error', n_jobs=-1)
grid_search.fit(X_train, y_train)
 
best_model = grid_search.best_estimator_
```

```python
best_model
```

```python
#cross validation: data split into 5, check cross validation : mean & SD

from sklearn.model_selection import cross_val_score
 
 
cross_val_scores = cross_val_score(best_model, X_train, y_train, cv=5, scoring='neg_mean_squared_error')
 
 
print("Cross-validated scores:", cross_val_scores)
 
 
print("Mean accuracy:", cross_val_scores.mean())
print("Standard deviation of accuracy:", cross_val_scores.std())
```

```python
#pickling : save all trained data

import joblib
 
joblib.dump(best_model, 'trained_model.joblib')
```

```python
#find accuracy

accuracy = best_model.score(X_test, y_test)
print('Accuracy = '+ str(accuracy))
```
